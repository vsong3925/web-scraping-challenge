```python
from splinter import Browser
from bs4 import BeautifulSoup
import pymongo
from webdriver_manager.chrome import ChromeDriverManager
```

```python
# setup splinter
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
conn = 'mongodb://localhost:27017'
client = pymongo.MongoClient(conn)
```

### 1) NASA Mars News

```python
db = client.mars_data
collection = db.news
```

```python
url = 'https://mars.nasa.gov/news/?page=0&per_page=40&order=publish_date+desc%2Ccreated_at+desc&search=&category=19%2C165%2C184%2C204&blank_scope=Latest'
```

```python
browser.visit(url)
```

```python
for x in range (1, 20):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    results = soup.find_all('li', class_='slide')
    
    for result in results:
        title = result.find('h3').text
        paragraph = result.find('div', class_='article_teaser_body').text
        
        # insert dictionary as a MongoDB document
        post = {
            'title': title,
            'paragraph': paragraph
        }
        
        collection.insert_one(post)
        
        print('page: ', x, '------------------')
        print(title)
        print(paragraph)
        
browser.links.find_by_partial_text('MORE').click()
```

### 2) JPL Mars Space Iamges - Featured Image

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Define MongoDB database and collection
db = client.mars_data
collection = db.images
```

```python
url = 'https://www.jpl.nasa.gov/spaceimages/?search=&category=Mars'
```

```python
browser.visit(url)
```

```python
for x in range (1, 40):
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    results = soup.find_all('li', class_='slide')
    
    for result in results:
        image = result.find('a', class_='fancybox')['data-fancybox-href']
        url = 'https://www.jpl.nasa.gov'
        image_url = f'{url}{image}'
        
        # Insert dictionary to MongoDB as a MongoDB document
        post = {
            'image_url': image_url
        }
        
        collection.insert_one(post)
        
        print('page: ', x, '-----------')
        print(image_url)

    browser.links.find_by_partial_text('MORE').click()
```

### 4) Mars Hemispheres

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# Define database and collections
db = client.mars_data
collection = db.hemispheres
```

```python
base_url = 'https://astrogeology.usgs.gov/search/map/Mars/Viking/'
hemispheres = ['cerberus', 'schiaparelli', 'syrtis_major', 'valles_marineris']
end_url = '_enhanced'

mars_list = []

for h in hemispheres:
    url = f'{base_url}{h}{end_url}'
    browser.visit(url)
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    
    results = soup.find_all('li')
    img_url = results[0].find('a')['href']
        #image = results[1].find('a')['href']
        
    mars_dict = {'title': f'{h} hemisphere', 'image_url': img_url}
    mars_list.append(mars_dict)
    
    collection.insert_one(mars_dict)
    
print(mars_list)
```
